```python
# Passo 1: Importar a Base de Dados
import pandas as pd

tabela = pd.read_csv("/content/drive/MyDrive/smartphones.csv")

# Passo 2: Visualizar e Tratar a Base de Dados
tabela = tabela.drop("Free", axis=1)

display(tabela)
```

```python
# Passo 3: Analisar as informações dos dados.

tabela = tabela.dropna()
display(tabela.info())
display(tabela.describe().round(2))
```

```python
qtde_marca = tabela["Brand"].value_counts()
display(qtde_marca)
```

```python
qtde_modelo = tabela["Model"].value_counts()
print(qtde_modelo)
print("-----------------------")
qtde_ram = tabela["RAM"].value_counts()
print(qtde_ram)
print("-----------------------")
qtde_storage = tabela["Storage"].value_counts()
print(qtde_storage)
print("-----------------------")
qtde_color = tabela["Color"].value_counts()
print(qtde_color)
```

```python
# Passo 4: Construir a Análise para identificar os principais motivos para a compra dos aparelhos.

import plotly.express as px

for coluna in tabela:
  grafico = px.histogram(tabela, x=coluna, color="Brand")

  grafico.show()
```

# Analise completa dos dados

## Dados Descobertos

- Preferências de Marca:

  - A maioria dos clientes prefere as marcas Samsung e Xiaomi.
Essas marcas têm mais vendas e predominância no mercado devido à maior variedade de aparelhos lançados em comparação com os concorrentes.

- Faixa de Preço:

  - A principal faixa de valor para compras está entre 150 a 250     dólares.

- Modelos Mais Vendidos:

  - Os dois modelos mais vendidos são Redmi Note 12 e Samsung S20.

- Preferência de Memória RAM:

  - A preferência dos clientes é por memória RAM entre 4 a 8 GB, com 8 GB sendo a mais vendida.

- Armazenamento:

  - A maior escolha de armazenamento é de 128 GB, predominando fortemente no mercado.


## Plano de Ação Estratégica

- Diversidade de Produtos:

  - As empresas que não vendem tantos aparelhos devem diversificar a linha de smartphones para alcançar todos os tipos de público.

- Cores de Smartphones:

  - Optar por cores mais escuras, como preto, azul ou cinza, tende a aumentar o interesse dos clientes.

- Armazenamento:

  - Lançar smartphones com armazenamento de 128 GB é uma excelente estratégia, já que essa capacidade é a mais escolhida pelos clientes.

- Memória RAM:

  - Smartphones com memória RAM abaixo de 4 GB e acima de 8 GB não têm uma alta taxa de vendas. Portanto, manter a oferta principalmente entre 4 e 8 GB é ideal.
- Faixa de Preço:

  - Considerando que a faixa de preço preferida é entre 150 a 250 dólares, é aconselhável evitar ultrapassar os 400 dólares para manter uma boa taxa de vendas.
